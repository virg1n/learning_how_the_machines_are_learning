```python
import torch
import matplotlib.pyplot as plt
import numpy as np
```

```python
names = open('names.txt').read().split('\n')
chars = {'.':0,}
```

```python
set_chars = set()
for i in names:
    for j in str(i):
        set_chars.add(j)
```

```python
for i, letter in enumerate(sorted(list(set_chars))):
    chars[letter] = i + 1
```

```python
list_of_names = '.'.join(names)
list_to_loop_of_names = ['.']
list_to_loop_of_names+= list(list_of_names)
keys_chars = list(chars.keys())
```

```python
N = torch.ones((27, 27), dtype=torch.int64)
for i in names[:]:
    list_to_loop_of_names = ['.'] + list(i) + ['.']
    for ch1, ch2 in list(zip(list_to_loop_of_names, list_to_loop_of_names[1:])):
        N[chars[ch1]][chars[ch2]] += 1
N[0][0] = 0
```

```python
array = N

plt.figure(figsize=(16, 16))  # Adjust size as needed

plt.imshow(array, cmap='viridis', interpolation='nearest')

plt.colorbar()

for (i, j), val in np.ndenumerate(array):
    plt.text(j, i, f'{val}\n{keys_chars[i]}{keys_chars[j]}', ha='center', va='center', color='white')


# Show the plot
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

```python
probs = torch.zeros((27, 27), dtype=torch.float64)
probs = N.to(torch.float64)
probs = N.to(torch.float64) / torch.sum(probs, dim=1, keepdim=True)
# probs
```

```python
probs[1]
```

```python
g = torch.Generator()
g.manual_seed(2147483647)
for i in range(15):
    name = '.'
    
    while True:
        name += str(keys_chars[torch.multinomial(probs[int(chars[name[-1]])], num_samples=1, replacement=False, generator=g).item()])
        if name[-1] == '.':
            print(name.replace('.', ''))
            break
```

```python
log_likelihood = 0
n = 0
for i in names[:3]:
    list_to_loop_of_names = ['.'] + list(i) + ['.']
    for ch1, ch2 in list(zip(list_to_loop_of_names, list_to_loop_of_names[1:])):
        n += 1
        prob = probs[chars[ch1]][chars[ch2]]
        log_prob = torch.log(prob)
        print(f'{ch1}{ch2}, prob: {prob}, log_prob={log_prob}')
        log_likelihood += log_prob
anll = -log_likelihood/n
anll
```

```python
xs, ys = [], []
for i in names[:1]:
    list_to_loop_of_names = ['.'] + list(i) + ['.']
    for ch1, ch2 in list(zip(list_to_loop_of_names, list_to_loop_of_names[1:])):
        xs.append(chars[ch1])
        ys.append(chars[ch2])

xs = torch.tensor(xs)
ys = torch.tensor(ys)
xs_one_hot = torch.nn.functional.one_hot(xs, num_classes=27).to(torch.float64)
ys_one_hot = torch.nn.functional.one_hot(ys, num_classes=27)
```

```python
g = torch.Generator()
g.manual_seed(2147483647)
W = torch.randn((27,27), requires_grad=True, generator=g, dtype=torch.float64)
```

```python
probs_xs = xs_one_hot @ W
logists = probs_xs.exp()
output = logists/ logists.sum(dim=1, keepdim=True)
loss = -sum(torch.log(output[torch.arange(5), ys]))/5
loss
```

```python
output
```

```python
W.grad = None
loss.backward()
```

```python
W.data += -0.1 * W.grad
```

```python
log_likelihood = 0
n = 0


g = torch.Generator()
g.manual_seed(2147483647)
W = torch.randn((27,27), requires_grad=True, generator=g, dtype=torch.float64)
xs, ys = [], []
for word in names:
    list_to_loop_of_names = ['.'] + list(word) + ['.']
    for ch1, ch2 in list(zip(list_to_loop_of_names, list_to_loop_of_names[1:])):
        xs.append(chars[ch1])
        ys.append(chars[ch2])
            
xs = torch.tensor(xs)
ys = torch.tensor(ys)
xs_one_hot = torch.nn.functional.one_hot(xs, num_classes=27).to(torch.float64)
ys_one_hot = torch.nn.functional.one_hot(ys, num_classes=27)
letter = xs.nelement()
```

```python
for epoch in range(100):
    probs_xs = xs_one_hot @ W
    logists = probs_xs.exp()
    output = logists/ logists.sum(dim=1, keepdim=True)
    loss = -output[torch.arange((letter)), ys].log().mean()
    # print(loss.item())
    W.grad = None
    loss.backward()
    W.data += -50 * W.grad
```
